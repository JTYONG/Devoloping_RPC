# Bottom Graphite

## Electron Signal

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the file - whitespace-separated, skip comment lines starting with #
df = pd.read_csv(
    "signal_BottomGraphite.txt",
    sep=r'\s+',
    comment='#',
    header=None,
    names=['TimeBin', 'Time_ns', 'ElectronSignal', 'IonSignal', 'TotalSignal']
)
plt.figure(figsize=(9, 5))
plt.plot(df['Time_ns'], df['ElectronSignal'], marker='o', linestyle='-', color='steelblue')
plt.xlabel('Time [ns]')
plt.ylabel('Electron Signal')
plt.title('RPC Induced Electron Signal at BottomGraphite')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Ion Signal

```python
plt.figure(figsize=(9, 5))
plt.plot(df['Time_ns'], df['IonSignal'], marker='o', linestyle='-', color='steelblue')
plt.xlabel('Time [ns]')
plt.ylabel('Electron Signal')
plt.title('RPC Induced Electron Signal at BottomGraphite')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
##Total Signal
```

```python
plt.figure(figsize=(9, 5))
plt.plot(df['Time_ns'], df["ElectronSignal"]+df['IonSignal'], marker='o', linestyle='-', color='steelblue')
plt.plot(df['Time_ns'], df['IonSignal'], marker='o', linestyle='-', color='red')
plt.xlabel('Time [ns]')
plt.ylabel('Electron Signal')
plt.title('RPC Induced Electron Signal at BottomGraphite')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Ion Signal

```python
df4 = pd.read_csv(
    "signal_BottomReadout.txt",
    sep=r'\s+',
    comment='#',
    header=None,
    names=['TimeBin', 'Time_ns', 'ElectronSignal', 'IonSignal', 'TotalSignal']
)
plt.figure(figsize=(9, 5))
plt.plot(df4['Time_ns'], df4['IonSignal'], marker='o', linestyle='-', color='steelblue')
plt.xlabel('Time [ns]')
plt.ylabel('Electron Signal')
plt.title('RPC Induced Electron Signal at Bottom Readout')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
