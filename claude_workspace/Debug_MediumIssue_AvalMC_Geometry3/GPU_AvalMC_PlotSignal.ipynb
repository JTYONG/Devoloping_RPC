```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# ── Load data ─────────────────────────────────────────────────────────────────
data = np.genfromtxt("Signal_TopReadout.csv", delimiter=",", skip_header=1, filling_values=0.0)

time    = data[:, 0]        # ns
prompt  = data[:, 1] * 1e9  # fC/ns → aC/ns
delayed = data[:, 2] * 1e9
total   = data[:, 3] * 1e9

# ── Plot ──────────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(time, -prompt,  color="#1f77b4", linewidth=1.8, marker="o", markersize=3.5, label="Total Prompt")
#ax.plot(time, delayed, color="#d62728", linewidth=1.8, marker="s", markersize=3.5, label="Total Delayed")
#ax.plot(time, -total,   color="#2ca02c", linewidth=2.0, marker="^", markersize=3.5, label="Total Signal")

ax.set_xlabel("Time [ns]", fontsize=13)
ax.set_ylabel("Cumulative Induced Charge [fC/ns]", fontsize=13)
ax.set_title("Cumulative Induced Charge", fontsize=15, fontweight="bold")

ax.grid(True, linestyle="--", linewidth=0.5, alpha=0.7)
ax.legend(fontsize=11)
ax.set_xlim(0,8)
ax.set_ylim(0,0.2)
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())
ax.tick_params(axis="both", which="major", labelsize=11)

plt.tight_layout()
plt.savefig("Signal_plot.png", dpi=150)
plt.show()
```
